```python
import pandas as pd
import matplotlib as pyplot
from datasets import load_dataset

#Load data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_ind = df[df['job_country'] == 'India'].copy()
df_ind['job_posted_month'] = df_ind['job_posted_date'].dt.strftime('%B')
df_ind_pivot = df_ind.pivot_table(index= 'job_posted_month', columns='job_title_short', aggfunc='size')
df_ind_pivot.reset_index(inplace=True)
df_ind_pivot['month_no'] = pd.to_datetime(df_ind_pivot['job_posted_month'], format= "%B").dt.month

df_ind_pivot.sort_values('month_no', inplace=True)
df_ind_pivot.set_index('job_posted_month',inplace=True)
df_ind_pivot.drop(columns='month_no',inplace=True)
df_ind_pivot
```

```python
df_ind_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')
df_ind_software_pivot
```

```python
df_ind_merge = df_ind_pivot.merge(df_ind_software_pivot, on="job_posted_month")
df_ind_merge
```

```python
top_5 = df_ind_merge.sum().sort_values(ascending=False).head().index.to_list()
top_5
```

```python
import matplotlib.pyplot as plt
df_ind_merge[top_5].plot(kind='line')

plt.title("Monthly Job Posting")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.ylim(0,20000)   # y-limit (0 to auto)
plt.legend()
plt.show()
```
